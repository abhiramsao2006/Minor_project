```python
#PROJECT 5: STUDENT MARKS ANALYZER
import numpy as np

# Accept number of students
n = int(input("Enter number of students: "))

# Accept marks
marks = []
for i in range(n):
    m = int(input(f"Enter marks of student {i+1}: "))
    marks.append(m)

# Store marks in NumPy array
marks_array = np.array(marks)

# Calculations
total_marks = np.sum(marks_array)
average_marks = np.mean(marks_array)
highest_marks = np.max(marks_array)
lowest_marks = np.min(marks_array)

# Grade logic using conditions
if average_marks > 85:
    grade = "A"
elif average_marks >=70:
    grade = "B"
elif average_marks >= 50:
    grade = "C"
else:
    grade = "Fail"

# Display results
print("--- Results--- ")
print("Total Marks:", total_marks)
print("Average Marks:", average_marks)
print("Highest Marks:", highest_marks)
print("Lowest Marks:", lowest_marks)
print("Grade:", grade)

#The program accepts marks for multiple students and calculates total marks,
#average marks, highest marks, and lowest marks. Based on the average marks, a grade is assigned using conditional statements.


#explanation
#The program first takes the number of students as input and then collects their marks in a list. The list is converted into a NumPy array to perform calculations easily



```

```python
#PROJECT 6: DAILY EXPENSE TRACKER
import numpy as np

# Input number of days
days = int(input("Enter number of days: "))

# Take daily expenses
expenses = []
for i in range(days):
    amount = float(input(f"Enter expense for day {i + 1}: "))
    expenses.append(amount)

# Store in NumPy array
expense_array = np.array(expenses)

# Calculations
total_expense = np.sum(expense_array)
average_expense = np.mean(expense_array)
highest_expense = np.max(expense_array)
highest_day = np.argmax(expense_array) + 1

# Display summary
print("\n--- Spending Summary ---")
print("Total Expense:", total_expense)
print("Average Expense:", average_expense)
print("Highest Expense:", highest_expense)
print("Highest Expense on Day:", highest_day)

# Optional warning
limit = 1000
if average_expense > limit:
    print("âš  Warning: Average expense exceeds the limit!")
else:
    print("Average expense is within the limit.")

  #The program records daily expenses and analyzes spending by calculating total expense, average expense, and highest expense. It also gives a warning if the average expense is high.


  #EXPLANATION
  #These expenses are stored in a list and then converted into a NumPy array. Using NumPy functions, the total, average, and highest expenses are calculated. The results are displayed, and a warning message is shown if the average expense exceeds a fixed limit.

```
